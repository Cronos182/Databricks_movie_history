### Spark Join Transformation

#### INNER JOIN

```python
%run "../includes/configuration"
```

```python
movie_df = spark.read.parquet(f"{silver_folder_path}/movies").filter("year_release_date = 2007")
```

```python
production_country_df = spark.read.parquet(f"{silver_folder_path}/productions_countries")
```

```python
country_df = spark.read.parquet(f"{silver_folder_path}/countries")
```

```python
display(movie_df)
```

```python
display(production_country_df)
```

```python
display(country_df)
```

```python
movie_production_country_df = movie_df.join(production_country_df, movie_df.movie_id == production_country_df.movie_id, "inner") \
.select(movie_df.title, movie_df.budget, production_country_df.country_id)
```

```python
display(movie_production_country_df)
```

```python
movie_country_df = movie_production_country_df.join(country_df, movie_production_country_df.country_id == country_df.country_id, "inner") \
    .select(movie_production_country_df.title, movie_production_country_df.budget, country_df.country_name)
```

```python
display(movie_country_df)
```

#### OUTER JOIN

##### Left Outer Join

```python
movie_production_country_df = movie_df.join(production_country_df, movie_df.movie_id == production_country_df.movie_id, "left") \
    .select(movie_df.title, movie_df.budget, production_country_df.country_id)
```

```python
display(movie_production_country_df)
```

```python
movie_country_df = movie_production_country_df.join(country_df, movie_production_country_df.country_id == country_df.country_id, "left") \
    .select(movie_production_country_df.title, movie_production_country_df.budget, country_df.country_name)
```

```python
display(movie_country_df)
```

##### Right Outer Join

```python
movie_production_country_df = movie_df.join(production_country_df, movie_df.movie_id == production_country_df.movie_id, "right") \
    .select(movie_df.title, movie_df.budget, production_country_df.country_id)
```

```python
display(movie_production_country_df)
```

```python
display(movie_country_df)
```

```python
movie_production_country_df = movie_df.join(production_country_df, movie_df.movie_id == production_country_df.movie_id, "full") \
    .select(movie_df.title, movie_df.budget, production_country_df.country_id)
```

##### Semi Join

```python
movie_production_country_df = movie_df.join(production_country_df, movie_df.movie_id == production_country_df.movie_id, "left") \
    .select(movie_df.title, movie_df.budget, production_country_df.country_id)
```

##### Anti Join

```python
movie_production_country_df = movie_df.join(production_country_df, movie_df.movie_id == production_country_df.movie_id, "left") \
    .select(movie_df.title, movie_df.budget, production_country_df.country_id)
```

```python
movie_country_df = movie_production_country_df.join(country_df, movie_production_country_df.country_id == country_df.country_id, "anti") \
    .select(movie_production_country_df.title, movie_production_country_df.budget)
```

```python
display(movie_country_df)
```

##### Cross Join

```python
movie_country_df = movie_df.crossJoin(country_df)
```
