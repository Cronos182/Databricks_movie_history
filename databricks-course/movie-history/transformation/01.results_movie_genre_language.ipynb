```python
movies_languages_df = spark.read.table("movie_silver.movies_languages") \
            .filter(f"file_date = '{v_file_date}'")
```

```python
# genres_df = spark.read.parquet(f"{silver_folder_path}/genres")
```

```python
genres_df = spark.read.table("movie_silver.genres")
```

```python
# movies_genres_df = spark.read.parquet(f"{silver_folder_path}/movies_genres")
```

```python
movies_genres_df = spark.read.table("movie_silver.movies_genres") \
            .filter(f"file_date = '{v_file_date}'")
```

#### Join "languages" y "movies_languages"

```python
# languages_mov_lan_df = languages_df.join(movies_languages_df, languages_df.language_id == movies_languages_df.language_id, "inner") \
# .select(languages_df.language_name, movies_languages_df.movie_id)    
```

```python
languages_mov_lan_df = languages_df.join(movies_languages_df, languages_df.language_id == movies_languages_df.language_id, "inner") \
.select(languages_df.language_name, languages_df.language_id, movies_languages_df.movie_id)
```

#### Join "genres" y "movies_genres"

```python
# genres_mov_gen_df = genres_df.join(movies_genres_df, genres_df.genre_id == movies_genres_df.genre_id, "inner") \
# .select(genres_df.genre_name, movies_genres_df.movie_id)
```

```python
genres_mov_gen_df = genres_df.join(movies_genres_df, genres_df.genre_id == movies_genres_df.genre_id, "inner") \
.select(genres_df.genre_name, genres_df.genre_id, movies_genres_df.movie_id)
```

#### Join "movies_df", "languages_mov_lan_df" y "genres_mov_lan_df"

- Filtrar las películas donde su fecha de lanzamiento sea mayor o igual a 2000

```python
movie_filter_df = movies_df.filter("year_release_date >= 2000")
```

```python
results_movies_genres_languages_df = movie_filter_df.join(languages_mov_lan_df, movie_filter_df.movie_id == languages_mov_lan_df.movie_id, "inner") \
.join(genres_mov_gen_df, movie_filter_df.movie_id == genres_mov_gen_df.movie_id, "inner")
```

```python
results_df = results_movies_genres_languages_df \
    .select(movie_filter_df.movie_id, "language_id", "genre_id","title", "duration_time", "release_date", "vote_average", "language_name", "genre_name") \
    .withColumn("created_date", lit(v_file_date))
```

- Ordenar por la columna "release_date" de manera descendente 

```python
results_order_by_df = results_df.orderBy(results_df.release_date.desc())
```
