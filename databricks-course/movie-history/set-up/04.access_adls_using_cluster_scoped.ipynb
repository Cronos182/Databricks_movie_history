### Acceder a Azure Data Lake Storage mediante Ámbito de Cluster
1. Establecer la configuración de spark "fs.azure.account.key" en el Cluster
2. Listar archivos del contenedor "demo"
3. Leer datos del archivo "movie.csv"

```python
dbutils.fs.ls("abfss://demo@moviehistory182.dfs.core.windows.net/")
```

Usar display para resultados en formato de tabla

```python
display(dbutils.fs.ls("abfss://demo@moviehistory182.dfs.core.windows.net/"))
```

API read de Spark

```python
display(spark.read.csv("abfss://demo@moviehistory182.dfs.core.windows.net/movie.csv"))
```
