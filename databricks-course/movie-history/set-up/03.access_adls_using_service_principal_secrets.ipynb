### Acceder a Azure Data Lake Storage mediante Service Principal
1. "Registrar la Aplicación" en Azure Entra ID/Service Principal
2. Generar un secreto(Contraseña) para la aplicación
3. Configurar Spark con APP/Client Id, Directory/Tenand Id & Secret
4. Asignar el Role "Storage Blob Data Contributor" al Data Lake

```python
client_id = dbutils.secrets.get(scope="movie-history-secret-scope", key="client-id")
tenant_id = dbutils.secrets.get(scope="movie-history-secret-scope", key="tenant-id")
client_secret = dbutils.secrets.get(scope="movie-history-secret-scope", key="client-secret")
```

```python
spark.conf.set("fs.azure.account.auth.type.moviehistory182.dfs.core.windows.net", "OAuth")
spark.conf.set("fs.azure.account.oauth.provider.type.moviehistory182.dfs.core.windows.net", "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider")
spark.conf.set("fs.azure.account.oauth2.client.id.moviehistory182.dfs.core.windows.net", client_id)
spark.conf.set("fs.azure.account.oauth2.client.secret.moviehistory182.dfs.core.windows.net", client_secret)
spark.conf.set("fs.azure.account.oauth2.client.endpoint.moviehistory182.dfs.core.windows.net", f"https://login.microsoftonline.com/{tenant_id}/oauth2/token")
```

```python
dbutils.fs.ls("abfss://demo@moviehistory182.dfs.core.windows.net/")
```
