### Acceder a Azure Data Lake Storage mediante Token SAS
1. Establecer la configuración de spark "SAS Token"
2. Listar archivos del contenedor "demo"
3. Leer datos del archivo "movie.csv"

```python
spark.conf.set("fs.azure.account.auth.type.moviehistory182.dfs.core.windows.net", "SAS")
spark.conf.set("fs.azure.sas.token.provider.type.moviehistory182.dfs.core.windows.net", "org.apache.hadoop.fs.azurebfs.sas.FixedSASTokenProvider")
spark.conf.set("fs.azure.sas.fixed.token.moviehistory182.dfs.core.windows.net", "sp=rl&st=2025-10-29T00:19:40Z&se=2025-10-29T08:34:40Z&spr=https&sv=2024-11-04&sr=c&sig=DfkEXQlcSulQmf3pB0kMzOooC5cO4JxEmlYb44YnapI%3D")
```

```python
dbutils.fs.ls("abfss://demo@moviehistory182.dfs.core.windows.net/")
```

Usar display para resultados en formato de tabla

```python
display(dbutils.fs.ls("abfss://demo@moviehistory182.dfs.core.windows.net/"))
```

API read de Spark

```python
display(spark.read.csv("abfss://demo@moviehistory182.dfs.core.windows.net/movie.csv"))
```
