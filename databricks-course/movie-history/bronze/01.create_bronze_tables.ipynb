### Crear tablas para archivos CSV

#### Crear la tabla "movies"

```sql
DROP TABLE IF EXISTS movie_bronze.movies;
CREATE TABLE IF NOT EXISTS movie_bronze.movies (
  movieId INT,
  title STRING,
  budget DOUBLE,
  homePage STRING,
  overview STRING,
  popularity DOUBLE,
  yearReleaseDate INT,
  releaseDate DATE,
  revenue DOUBLE,
  durationTime INT,
  movieStatus STRING,
  tagline STRING,
  voteAverage DOUBLE,
  voteCount INT
)
USING CSV
OPTIONS(path "abfss://bronze@moviehistory182.dfs.core.windows.net/movie.csv", header true)
```

```sql
SELECT * FROM movie_bronze.movies;
```

```sql
DROP TABLE IF EXISTS movie_bronze.languages;
CREATE TABLE IF NOT EXISTS movie_bronze.languages(
    languageId INT,
    languageCode STRING,
    languageName STRING
)
USING CSV
OPTIONS(path "abfss://bronze@moviehistory182.dfs.core.windows.net/language.csv", header true)
```

```sql
SELECT * FROM movie_bronze.languages;
```

```sql
DROP TABLE IF EXISTS movie_bronze.genres;
CREATE TABLE IF NOT EXISTS movie_bronze.genres(
    genreId INT,
    genreName STRING
)
USING CSV
OPTIONS(path "abfss://bronze@moviehistory182.dfs.core.windows.net/genre.csv", header true)
```
